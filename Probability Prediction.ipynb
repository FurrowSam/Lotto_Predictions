```python
lotto_df
```

```python
# Calculate frequency for all days (or melted DataFrame for overall probability)
number_counts_A = lotto_df['A'].value_counts()
# Normalize by total count to get probabilities
total_count_A = len(number_counts_A)
number_probabilities_A = number_counts_A / total_count_A
```

```python
number_probabilities_A
```

```python
number_counts_B = lotto_df['B'].value_counts()
# Normalize by total count to get probabilities
total_count_B = len(number_counts_B)
number_probabilities_B = number_counts_B / total_count_B
```

```python
number_counts_C = lotto_df['C'].value_counts()
# Normalize by total count to get probabilities
total_count_C = len(number_counts_C)
number_probabilities_C = number_counts_C / total_count_C
```

```python
number_counts_D = lotto_df['D'].value_counts()
# Normalize by total count to get probabilities
total_count_D = len(number_counts_D)
number_probabilities_D = number_counts_D / total_count_D
```

```python
number_counts_E = lotto_df['E'].value_counts()
# Normalize by total count to get probabilities
total_count_E = len(number_counts_E)
number_probabilities_E = number_counts_E / total_count_E
```

```python
number_counts_F = lotto_df['F'].value_counts()
# Normalize by total count to get probabilities
total_count_F = len(number_counts_F)
number_probabilities_F = number_counts_F / total_count_F
```

```python
lotto_df = pd.concat([number_probabilities_A, number_probabilities_B, number_probabilities_C, number_probabilities_D, number_probabilities_E, number_probabilities_F])
```

```python
lotto_df = lotto_df.to_frame()
```

```python
lotto_df.sort_index()
```

```python

lotto_grouped_df = lotto_df.groupby(lotto_df.index)

# Use 'sum' to aggregate the values for each group (1's, 2's, etc.)
lotto_combined_df = lotto_grouped_df.sum()

# Print the DataFrame with combined values
print(lotto_combined_df)
```

```python
lotto_final = lotto_combined_df.sort_values('count', ascending=True)
```
